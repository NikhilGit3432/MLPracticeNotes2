```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Logistic Regression 
is a statistical and machine learning algorithm used for binary classification (i.e., classifying data into two categories, such as Yes/No, Spam/Not Spam, 0/1). It predicts the probability that a given input belongs to a particular class.

```python
from sklearn.datasets import make_classification
```

```python
# Define parameters
n_samples = 1000
n_features = 10
n_redundant = 5
n_informative = 5
n_classes = 2
random_state = 1
```

```python
# Generate dataset
X, y = make_classification(n_samples=n_samples, n_features=n_features, 
                           n_redundant=n_redundant, n_informative=n_informative, 
                           n_classes=n_classes, random_state=random_state)

# Convert to DataFrame
df = pd.DataFrame(X, columns=[f'Feature_{i+1}' for i in range(n_features)])
df['Target'] = y

# Display first few rows
df
```

# 3:38
# Splitting the data into Train and Test splits

```python
# Split the dataset into train and test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

# Importing Logistic Regression Machne Learning Model
- Since the datset we are dealing with is a Calssification type we will need Logistics regression Model to train it.

```python
# Create an instance of Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Train the model on the training data
model.fit(X_train, y_train)
```

# Determine the values for y_pred

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

# 5:39 evluation Metrices 

```python
# Evaluate the model's performance
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)
```

# 7:15 ROC AUC Score

```python
# First Prdict the y_pred_proba
# Get predicted probabilities for the test set
y_pred_proba = model.predict_proba(X_test)

# Display the predicted probabilities
y_pred_proba
```

```python
# Now detrmine fpr,tpr and threshold
from sklearn.metrics import roc_curve

# Get predicted probabilities for the positive class (class 1)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calculate TPR, FPR, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Display the results
print("False Positive Rate (FPR):", fpr)
print("True Positive Rate (TPR):", tpr)
print("Thresholds:", thresholds)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
# Calculate AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python

from sklearn.metrics import precision_score, recall_score, accuracy_score


# Get predicted probabilities for the positive class (class 1)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Define 100 thresholds from 0 to 1
thresholds = np.linspace(0, 1, 100)

# Initialize lists to store precision, recall, and accuracy scores
precision_scores = []
recall_scores = []
accuracy_scores = []

# Calculate precision, recall, and accuracy for each threshold
for threshold in thresholds:
    # Convert probabilities to binary predictions based on the threshold
    y_pred = (y_pred_proba >= threshold).astype(int)
    
    # Calculate precision, recall, and accuracy
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Append scores to the lists
    precision_scores.append(precision)
    recall_scores.append(recall)
    accuracy_scores.append(accuracy)

# Display the results
print("Thresholds:", thresholds)
print("Precision Scores:", precision_scores)
print("Recall Scores:", recall_scores)
print("Accuracy Scores:", accuracy_scores)
```

```python
import matplotlib.pyplot as plt

# Plot Precision, Recall, and Accuracy vs Thresholds
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precision_scores, label='Precision', color='blue', lw=2)
plt.plot(thresholds, recall_scores, label='Recall', color='green', lw=2)
plt.plot(thresholds, accuracy_scores, label='Accuracy', color='red', lw=2)
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision, Recall, and Accuracy vs Thresholds')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

### Conclusion:

#### Accuracy Improved Slightly

- Before setting the cutoff to 0.4: Accuracy = 0.8250 (82.5%)
- After setting the cutoff to 0.4: Accuracy = 0.8300 (83.0%)
- The accuracy has improved slightly, indicating that adjusting the threshold helped classify more samples correctly.

#### Reduction in False Negatives (Missed Positives)

- Before: False Negatives = 25 (class 1 misclassified as 0)
- After: False Negatives = 17
- Lowering the threshold to 0.4 reduced the number of false negatives, meaning more actual positives were correctly classified.

#### Increase in False Positives

- Before: False Positives = 10 (class 0 misclassified as 1)
- After: False Positives = 17
- More negative samples (class 0) were misclassified as positive.


#### Better Balance Between Precision and Recall

- Before: Class 1 had high precision (0.89) but lower recall (0.76), meaning it was conservative in predicting positives.
- After: Class 1 had a more balanced precision (0.84) and recall (0.84), meaning it captured more positives at the cost of slightly lower precision.


- Since accuracy remained nearly the same, this trade-off may be acceptable depending on the application.

- If false negatives are more costly (e.g., in medical diagnosis), the new threshold is better.
- If false positives are more costly (e.g., fraud detection), the original threshold might be preferable.
