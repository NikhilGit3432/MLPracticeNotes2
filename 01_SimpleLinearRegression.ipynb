# Regression 
In machine learning, regression refers to a type of supervised learning task where the goal is to predict a continuous output variable based on input features. It involves establishing a relationship between the input variables (independent variables) and the output variable (dependent variable) so that we can use the model to make predictions on unseen data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For this we are using Diabities dataset
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes.keys()
```

```python
df_diabetes = pd.DataFrame(diabetes.data, columns = [diabetes.feature_names])
df_diabetes["Target"] = diabetes.target
```

```python
df_diabetes
```

### Simple Linear Regression (SLR) in Short:
Simple Linear Regression is a statistical method used to model the relationship between two variables:

- Independent variable (X): The predictor or input feature.
- Dependent variable (y): The target or output.

```python
# Choose a feature for visualization (e.g., "bmi")
feature = "bmi"
target = "Target"



# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df_diabetes[feature], df_diabetes[target], color='blue', alpha=0.7, label=f"{feature} vs {target}")

# Add labels, title, and legend
plt.title(f"Scatter Plot: {feature} vs {target}", fontsize=16)
plt.xlabel(f"{feature.capitalize()}", fontsize=14)
plt.ylabel("Target", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

## Creating a Machuine Learning Model 
Since we are going to practice simple linear regression machine learning model we have to choose only one Feature nand one Traget Variable 

```python
## Divide the Datainto X,y
X = df_diabetes["bmi"] # As input or independent feature 
y = df_diabetes["Target"] # As dependent or target varaible 
```

## 16.45

```python
## Split the data into train and test
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=1)
```

## 22.03

```python
## Scalling the data (Optional)
```

```python
## Model training 
from sklearn.linear_model import LinearRegression
```

## 24.40

```python
# Instanceiate the model
Lr_model = LinearRegression()
Lr_model
```

## 25.30

```python
## Training the model
Lr_model.fit(X_train,y_train) # at backend it will use gradient descent to determine the optimal values of theeta zero, theeta one
```

## 30.30


```python
# Retrieve the slope (coefficient) of the linear regression model
# It represents the change in the target variable for a unit change in the independent variable (BMI in this case).

Lr_model.coef_ 
```

```python
# Retrieve the intercept of the linear regression model
# It represents the predicted value of the target variable when the independent variable (BMI in this case) is zero.

Lr_model.intercept_
```

## 31.24

```python
# Predict the target values (y) for the test dataset (X_test)
# The model uses the learned slope (coef_) and intercept (intercept_) to compute predictions.

y_pred =  Lr_model.predict(X_test)
y_pred
```

## 33:56

```python
import matplotlib.pyplot as plt

# Scatter plot for the actual data
plt.scatter(X_test, y_test, color="blue", label="Actual Data", alpha=0.7)

# Plot the linear regression line
plt.plot(X_test, y_pred, color="red", label="Regression Line", linewidth=2)

# Add labels, title, and legend
plt.title("Linear Regression: Actual Data vs Prediction", fontsize=16)
plt.xlabel("BMI (X_test)", fontsize=14)
plt.ylabel("Target (y)", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Display the plot
plt.show()
```
